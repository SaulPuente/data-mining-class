## Electrificacionrural.csv

```python
df = pd.read_csv("data/Electrificacionrural.csv")
df
```

```python
df = df.drop([0, 1, 2, 37, 38, 39, 40, 41, 42]).drop(["Unnamed: 16"], axis=1)
df = df.reset_index().drop(["index"], axis=1)
df.head()
```

```python
estados_lista = df["COMISION FEDERAL DE ELECTRICIDAD"]
estados_lista = estados_lista.reset_index().drop([0]).reset_index().drop(["index", "level_0"], axis=1)
estados_lista.rename({"COMISION FEDERAL DE ELECTRICIDAD": 'ESTADO'}, axis=1, inplace=True)
estados_lista.head()
```

```python
df = df.drop(["COMISION FEDERAL DE ELECTRICIDAD"], axis=1)
df = df.reset_index().drop(["index"], axis=1).set_axis(range(df.shape[1]), axis=1)
df.head()
```

```python
periodos_lista = df.iloc[0]
periodos_lista.head()
```

```python
df = df.drop([0]).reset_index().drop(["index"], axis=1)
df.head()
```

```python
dict_df = dict()
dict_df["estado"] = list()
dict_df["periodo"] = list()
dict_df["electrificacion"] = list()
for i in range(df.shape[0]):
    for j in range(df.shape[1]-4):
        dict_df["estado"].append(estados_lista["ESTADO"].iloc[i])
        dict_df["periodo"].append(periodos_lista.iloc[j])
        dict_df["electrificacion"].append(df.iloc[i][j])
    dict_df["estado"].append(estados_lista["ESTADO"].iloc[i])
    dict_df["periodo"].append("2021")
    dict_df["electrificacion"].append(df.iloc[i][df.shape[1]-1])
```

```python
dict_df
```

```python
clean_df = pd.DataFrame.from_dict(dict_df)
clean_df.head()
```

```python
clean_df.to_csv("data/Electrificacionrural_clean.csv", index=False)
```

## Usuariosyconsumodeelectricidadpormunicipio.csv

```python
df = pd.read_csv("data/Usuariosyconsumodeelectricidadpormunicipio.csv")
df
```

```python
df.columns = df.iloc[1]
df
```

```python
m = df.index[df["Cve Edo"] == "CONSUMO 2010A  2017 UNIDAD EN (KW/H)"].tolist()[0]
m
```

```python
usuarios_df = df.loc[:m,:]
usuarios_df.head()
```

```python
consumo_df = df.loc[m:,:]
consumo_df.head()
```

```python
usuarios_df = usuarios_df.drop([0, 1])
usuarios_df = usuarios_df.reset_index().drop(["index"], axis=1)
usuarios_df.index.name = 'index'
usuarios_df = usuarios_df.melt(id_vars=list(usuarios_df.columns[:-8]), 
        var_name="Periodo", 
        value_name="Usuarios")
usuarios_df.head()
```

```python
consumo_df = consumo_df.drop([112144])
consumo_df = consumo_df.reset_index().drop(["index"], axis=1)
consumo_df.index.name = 'index'
consumo_df = consumo_df.melt(id_vars=list(consumo_df.columns[:-8]), 
        var_name="Periodo", 
        value_name="Consumo")
consumo_df.head()
```

```python
df = df.drop([0, 1])
df = df.reset_index().drop(["index"], axis=1)
df.index.name = 'index'
df = df.melt(id_vars=list(df.columns[:-8]), 
        var_name="Periodo", 
        value_name="Cantidad")
df.head()
```

```python
usuarios_consumo_df = usuarios_df
usuarios_consumo_df["Consumo"] = consumo_df["Consumo"]
usuarios_consumo_df.head()
```

```python
del df, usuarios_df, consumo_df
```

```python
usuarios_consumo_df.info()
```

```python
usuarios_consumo_df['Usuarios'] = usuarios_consumo_df['Usuarios'].replace(',','', regex=True)
usuarios_consumo_df['Usuarios'] = usuarios_consumo_df['Usuarios'].astype(float)

usuarios_consumo_df['Consumo'] = usuarios_consumo_df['Consumo'].replace(',','', regex=True)
usuarios_consumo_df['Consumo'] = usuarios_consumo_df['Consumo'].astype(float)
```

```python
usuarios_consumo_df['Entidad Federativa'].unique()
```

```python
usuarios_consumo_df['Entidad Federativa'] = usuarios_consumo_df['Entidad Federativa'].str.strip()
usuarios_consumo_df['Entidad Federativa'] = usuarios_consumo_df['Entidad Federativa'].apply(lambda x: x if x != "N a c i o n a l" else x.replace(' ',''))
usuarios_consumo_df['Municipio'] = usuarios_consumo_df['Municipio'].str.strip()
usuarios_consumo_df['Cve Edo'] = usuarios_consumo_df['Cve Edo'].str.strip()
usuarios_consumo_df['Cve Inegi'] = usuarios_consumo_df['Cve Inegi'].str.strip()
usuarios_consumo_df['Tarifa'] = usuarios_consumo_df['Tarifa'].str.strip()
```
