```python
import pandas as pd
```

```python
df = pd.read_csv("data/Usuariosyconsumodeelectricidadpormunicipio_clean.csv", dtype={
    'Cve Edo': 'string',
    'Cve Inegi': 'string',
    'Cve Mun': 'string',
    'Entidad Federativa': 'string',
    'Municipio': 'string',
    'Tarifa': 'string',
    'Periodo': 'int64',
    'Consumo': 'float64'
})
df.head()
```

```python
df[["Consumo"]].describe()
```

```python
df[["Entidad Federativa", "Consumo"]].groupby(by="Entidad Federativa").sum().sort_values(by='Consumo', ascending=False)
```

```python
df[["Entidad Federativa", "Periodo", "Consumo"]]\
.groupby(by=["Entidad Federativa", "Periodo"]).sum().sort_values(by=["Entidad Federativa", "Periodo"], ascending=True)
```

```python
df[["Entidad Federativa", "Municipio", "Consumo"]].loc[df["Entidad Federativa"] == "Nuevo Leon"]\
.groupby(by=["Entidad Federativa", "Municipio"]).sum().sort_values(by='Consumo', ascending=False)
```
