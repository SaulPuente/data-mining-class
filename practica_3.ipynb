```python
import pandas as pd
```

```python
df = pd.read_csv("data/Usuariosyconsumodeelectricidadpormunicipio_clean.csv", dtype={
    'Cve_Edo': 'string',
    'Cve_Inegi': 'string',
    'Cve_Mun': 'string',
    'Estado': 'string',
    'Municipio': 'string',
    'Tarifa': 'string',
    'Periodo': 'int64',
    'Usuarios': 'int64',
    'Consumo': 'float64',
    'ISO': 'string'
})
df.head()
```

```python
df[["Usuarios", "Consumo"]].describe()
```

```python
df[["Estado", "Consumo", "Usuarios"]].groupby(by="Estado").sum().sort_values(by='Consumo', ascending=False)
```

```python
df[["Estado", "Periodo", "Consumo", "Usuarios"]]\
.groupby(by=["Estado", "Periodo"]).sum().sort_values(by=["Estado", "Periodo"], ascending=True)
```

```python
df[["Estado", "Municipio", "Consumo", "Usuarios"]].loc[df["Estado"] == "Nuevo Leon"]\
.groupby(by=["Estado", "Municipio"]).sum().sort_values(by='Consumo', ascending=False)
```
